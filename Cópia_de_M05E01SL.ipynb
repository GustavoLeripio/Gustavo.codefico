6 - Pergunta: Explique a diferença entre elementos de bloco e elementos inline em HTML. Dê exemplos de cada um.

blocos:
elementos de bloco são usados para estruturar o conteúdo da página em blocos distintos, eles normalmente começam em uma nova linha e ocupam toda a largura disponível do elemento pai que os contém.

inlines:
elementos inline são usados para marcar partes de texto ou conteúdo que ocorrem dentro de elementos de bloco, eles não começam em uma nova linha e ocupam apenas o espaço necessário para o conteúdo que envolvem.
