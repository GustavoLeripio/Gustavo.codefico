Nesta lista de exercícios, você terá a oportunidade de práticar os conceitos da aula sobre seções. Nesta lista, iremos criar uma página de uma agência de turismo e essa agência têm três destinos que ela presta serviços. No exercício proposto iremos utilizar elementos que aprendemos para organizar nossa página por partes. Siga as instruções cuidadosamente.

Instruções:

1. Título: Comece criando um documento HTML válido com uma estrutura básica, incluindo a declaração <!DOCTYPE html>, as tags \<html>, \<head> e \<body>. Defina um título apropriado para a página.

2. Definir parte: No \<body>, defina uma parte na página usando a tag \<div>.

3. Cabeçalho: Na \<div>, crie um cabeçalho \<header> com o nome da agência (\<h1>) e uma breve descrição (\<p>).

4. Destinos: Na \<div>, abaixo do cabeçalho, implemente três seções (\<section>), cada uma representando um destino de viagem, use o atributo id = "nome_do_destino" na tag \<section> para identificar cada destino. Inclua em cada seção um título (\<h2>) com nome do destino e uma página externa incorporada utlizando a tag \<iframe>, essa página pode ser o wikipedia do destino. Acesse o wikipedia, pesquise o destino, copie e cola a url na tag \<iframe>, utilize os atributos height="400" width="600" na tag \<iframe> para ajustar o tamanho.

5. Barra de navegação: Na \<div>, depois do\<header>, você irá inserir a tag \<nav>, com links para cada seção. Como foi visto em aula, vocês utilizarão a tag \<a> para cada seção. Em "href" vocês irão atribuir o "id" de cada seção, por exemplo: \<a href="#cancun">Cancún \</a>.

6. Rodapé: No final da \<div> inclua um rodapé (\<footer>) contendo um parágrafo (\<p>) com o nome do autor da página, por exemplo: "\<p> Página criada por Lucas Leite \</p>"

* Os destinos e informações vocês podem escolher.

```python
<!DOCTYPE html>
<html>
<head>
    <title>Agência de Viagens</title>
</head>
<body>
    <div>
        <header>
            <h1>Viagens</h1>
            <p>Sua escolha para aventuras pelo mundo!</p>
        </header>

        <nav>
            <a href="#FernandodeNoronha">Fernando De Noronha</a>
            <a href="#SantaCatarina">Santa Catarina</a>
            <a href="#FozDoIguaçu">Foz Do Iguaçu</a>
        </nav>

        <section id="FernandodeNoronha">
            <h2>Fernando De Noronha</h2>
            <iframe src="https://pt.wikipedia.org/wiki/Fernando_de_Noronha" width="600" height="400"></iframe>
        </section>

        <section id="#SantaCatarina">
            <h2>Santa Catarina</h2>
            <iframe src="https://pt.wikipedia.org/wiki/Santa_Catarina" width="600" height="400"></iframe>
        </section>

        <section id="Fozdoiguaçu">
            <h2>Foz Do Iguaçu</h2>
            <iframe src="https://pt.wikipedia.org/wiki/Foz_do_Igua%C3%A7u" width="600" height="400"></iframe>
        </section>

        <footer>
            <p>Página criada por Gustavo Leripio</p>
        </footer>
    </div>
</body>
</html>
```
