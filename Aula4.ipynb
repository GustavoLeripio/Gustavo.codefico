Nesta lista de exerccios, voc ter a oportunidade de prticar os conceitos da aula sobre sees. Nesta lista, iremos criar uma pgina de uma agncia de turismo e essa agncia tm trs destinos que ela presta servios. No exerccio proposto iremos utilizar elementos que aprendemos para organizar nossa pgina por partes. Siga as instrues cuidadosamente.

Instrues:

1. Ttulo: Comece criando um documento HTML vlido com uma estrutura bsica, incluindo a declarao <!DOCTYPE html>, as tags \<html>, \<head> e \<body>. Defina um ttulo apropriado para a pgina.

2. Definir parte: No \<body>, defina uma parte na pgina usando a tag \<div>.

3. Cabealho: Na \<div>, crie um cabealho \<header> com o nome da agncia (\<h1>) e uma breve descrio (\<p>).

4. Destinos: Na \<div>, abaixo do cabealho, implemente trs sees (\<section>), cada uma representando um destino de viagem, use o atributo id = "nome_do_destino" na tag \<section> para identificar cada destino. Inclua em cada seo um ttulo (\<h2>) com nome do destino e uma pgina externa incorporada utlizando a tag \<iframe>, essa pgina pode ser o wikipedia do destino. Acesse o wikipedia, pesquise o destino, copie e cola a url na tag \<iframe>, utilize os atributos height="400" width="600" na tag \<iframe> para ajustar o tamanho.

5. Barra de navegao: Na \<div>, depois do\<header>, voc ir inserir a tag \<nav>, com links para cada seo. Como foi visto em aula, vocs utilizaro a tag \<a> para cada seo. Em "href" vocs iro atribuir o "id" de cada seo, por exemplo: \<a href="#cancun">Cancn \</a>.

6. Rodap: No final da \<div> inclua um rodap (\<footer>) contendo um pargrafo (\<p>) com o nome do autor da pgina, por exemplo: "\<p> Pgina criada por Lucas Leite \</p>"

* Os destinos e informaes vocs podem escolher.

```python
(<!DOCTYPE html>
<html>
<head>
    <title>Agncia de Viagens</title>
</head>
<body>
    <div>
        <header>
            <h1>Viagens</h1>
            <p>Sua escolha para aventuras pelo mundo!</p>
        </header>

        <nav>
            <a href="#FernandodeNoronha">Fernando De Noronha</a>
            <a href="#SantaCatarina">Santa Catarina</a>
            <a href="#FozDoIguau">Foz Do Iguau</a>
        </nav>

        <section id="FernandodeNoronha">
            <h2>Fernando De Noronha</h2>
            <iframe src="https://pt.wikipedia.org/wiki/Fernando_de_Noronha" width="600" height="400"></iframe>
        </section>

        <section id="#SantaCatarina">
            <h2>Santa Catarina</h2>
            <iframe src="https://pt.wikipedia.org/wiki/Santa_Catarina" width="600" height="400"></iframe>
        </section>

        <section id="Fozdoiguau">
            <h2>Foz Do Iguau</h2>
            <iframe src="https://pt.wikipedia.org/wiki/Foz_do_Igua%C3%A7u" width="600" height="400"></iframe>
        </section>

        <footer>
            <p>Pgina criada por Gustavo Leripio</p>
        </footer>
    </div>
</body>
</html>)
```
