Nesta lista de exercícios, você terá a oportunidade de criar um formulário HTML detalhado. Siga as instruções cuidadosamente para criar um formulário funcional e bem estruturado.

Instruções:

1. Título: Comece criando um documento HTML válido com uma estrutura básica, incluindo a declaração <!DOCTYPE html>, as tags <html>, <head> e <body>. Defina um título apropriado para a página.

2. Cabeçalho: No \<body>, crie um cabeçalho (\<h1>) que descreva o propósito do formulário (por exemplo, "Formulário de Contato").

3. Formulário: Abaixo do cabeçalho, inicie a criação do formulário. Use a tag \<form> para envolver todo o formulário.

4. Campos de Entrada de Dados: <br/>
a. Crie um campo de entrada de texto para o nome do usuário. Inclua um \<label> associado a este campo.<br/>
b. Crie um campo de entrada de e-mail para o endereço de e-mail do usuário. Inclua um \<label> associado a este campo e defina o atributo type como "email".
c. Crie um campo de senha para a senha do usuário. Inclua um \<label> associado a este campo e defina o atributo type como "password".<br/>
d. Adicione um campo de seleção de gênero. Use a tag \<select> com as opções "Masculino" e "Feminino". Inclua um \<label> associado a este campo.<br/>

5. Campos de Opção:<br/>
a. Crie uma pergunta de múltipla escolha com três opções. Use a tag \<fieldset> para agrupar as opções e inclua um \<legend> que descreva a pergunta.<br/>
b. Crie três opções de resposta usando \<input type="radio">. Certifique-se de vincular cada opção a uma descrição usando \<label>. Todas as opções devem ter o mesmo atributo name.<br/>

6. Área de Texto: Crie uma área de texto (\<textarea>) para que os usuários possam escrever um comentário ou mensagem adicional. Inclua um \<label> associado a este campo.

7. Botões:
a. Adicione um botão de envio (\<input type="submit">) para enviar o formulário.<br/>
b. Adicione um botão de redefinição (\<input type="reset">) para limpar o formulário.<br/>

8. Dica de Usabilidade: Use o atributo placeholder para fornecer dicas sobre o que deve ser inserido em cada campo de entrada.<br/>

9. Validação: Utilize o atributo required em pelo menos um dos campos para indicar que é obrigatório preenchê-lo.

10. (Extra) Pesquise diferentes valores do atributo type no elemento \<input> e os utilize. Comece criando campos de entrada específicos para números, faixa de números e URLs, utilizando os tipos apropriados.

```python
Codigo


<!DOCTYPE html>
<html>
<head>
    <title>Formulário de Contato</title>
</head>
<body>
    <h1>Formulário de Contato</h1>

    <form>
        <label for="nome">Nome:</label>
        <input type="text" id="nome" name="nome" placeholder="Digite seu nome" required><br><br>

        <label for="email">E-mail:</label>
        <input type="email" id="email" name="email" placeholder="Digite seu e-mail" required><br><br>

        <label for="senha">Senha:</label>
        <input type="password" id="senha" name="senha" placeholder="Digite sua senha" required><br><br>

        <label for="genero">Gênero:</label>
        <select id="genero" name="genero">
            <option value="masculino">Masculino</option>
            <option value="feminino">Feminino</option>
        </select><br><br>
        <fieldset>
            <legend>Pergunta de Multipla Escolha</legend>
            <label for="opcao1">
                <input type="radio" id="opcao1" name="escolha 1" value="opcao1">Opção 1
            </label><br>
            <label for="opcao2">
                <input type="radio" id="opcao2" name="escolha 2" value="opcao2">Opção 2
            </label><br>
            <label for="opcao3">
                <input type="radio" id="opcao3" name="escolha 3" value="opcao3">Opção 3
            </label><br>
        </fieldset><br><br>

        <label for="numero">Número:</label>
            <input type="number" id="numero" name="numero"><br><br>

        <label for="comentario">Comentário:</label><br>
        <textarea id="comentario" name="comentario" placeholder="Digite seu comentário"></textarea><br><br>

        <label for="faixa">Faixa de números:</label>
        <input type="range" name="faixa" id="faixa" min="0" max="150" /><br><br>

        <input type="submit" value="Enviar">
        <input type="reset" value="Limpar">

    </fieldset>
    </form>
</body>
</html>
```
